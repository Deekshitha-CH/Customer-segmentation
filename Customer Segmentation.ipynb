CREATING A COLUMN TO SHOW WHETHER IN A RELATION OR NOT

```python
def get_relationship(row):
    if 'marital_married' in row and row['marital_married'] == 1:
        return 1
    elif 'marital_Together' in row and row['marital_Together'] == 1:
        return 1
    else:
        return 0
```

```python
data['In_relationship'] = data.apply(get_relationship, axis=1)
data.head()
```

CORRECTING A COLUMN NAME

```python
data.rename(columns={'education_2n Cycle\t': 'education_2n Cycle'}, inplace=True)
```

ASSIGNING EDUCATION STATUS TO DIFFERENT ROWS

```python
def Education_Status(row):
    if row['education_2n Cycle'] == 1:
        return '2'
    elif row['education_Basic'] == 1:
        return '1'
    elif row['education_Graduation'] == 1:
        return '3'
    elif row['education_Master'] == 1:
        return '4'
    elif row['education_PhD'] == 1:
        return '5'
    else:
        return 'Unknown'

```

```python
data['Education_status'] = data.apply(Education_Status, axis=1)
```

SEGMENTATION

IMPORTING K MEANS

```python
from sklearn.cluster import KMeans
```

STANDARDISING THE SCALE OF DIFFERENT COLUMNS TO BE USED IN SEGMENTATION

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
cols_for_clustering = ['Income', 'MntTotal', 'In_relationship','Education_status']
data_scaled = data.copy()
data_scaled[cols_for_clustering] = scaler.fit_transform(data[cols_for_clustering])
```

CHOOSING A RANGE FOR CLUSTER VALUES FOR ELBOW METHOD

```python
X = data_scaled[cols_for_clustering]
inertia_list = []
for k in range(1,11):
    inertia = KMeans(n_clusters=k , n_init=10).fit(X).inertia_
    inertia_list.append(inertia)
```

VISUALIZATION OF ELBOW METHOD PLOT

```python
plt.figure(figsize=[7,5])
plt.plot(range(1,11), inertia_list) 
plt.title("Inertia vs. Number of Clusters")
plt.xlabel("Number of Clusters (K)")
plt.ylabel("Inertia")
plt.show()
```

FITTING DESIRED NUMBER OF CLUSTERS INTO OUR MODEL AND CREATING SEGMENTS

```python
model = KMeans(n_clusters=4, random_state = 7)
model.fit(data_scaled[cols_for_clustering])
data_scaled['Cluster'] = model.predict(data_scaled[cols_for_clustering])
```

GENERATING 2 MOST RELAVENT PROPERTIES OF CLUSTERS

```python
from sklearn import decomposition
pca= decomposition.PCA(n_components= 2)
pca_res = pca.fit_transform(data_scaled[cols_for_clustering])
data_scaled['pc1']= pca_res[:,0]
data_scaled['pc2']= pca_res[:,1]
```

VISUALIZATION OF THE CLUSTERS

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x='pc1', y='pc2', data=data_scaled, hue='Cluster', palette='viridis')
plt.title('Clustered Data Visualization')
plt.xlabel('Principal Component 1 (pc1)')
plt.ylabel('Principal Component 2 (pc2)')
plt.legend(title='Clusters')
```

PROPERTIES OF CLUSTERS

NUMBER OF CUSTOMERS PER CLUSTER AND ITS VISUALIZATION

```python
customer_num = data_scaled.groupby('Cluster')['MntTotal'].count().reset_index()

plt.figure(figsize=(8, 6))
sns.barplot(x='Cluster', y='MntTotal', data=customer_num, palette='viridis')
plt.title('Cluster sizes')
plt.xlabel('Cluster')
plt.ylabel('customer_num')
plt.show()
```

FORMING CLUSTERS WITHOUT MODIFYING THE ORIGINAL VALUES OF COLUMNS USED TO FORM CLUSTERS

```python
scaler = StandardScaler()
cols_for_clustering = ['Income', 'MntTotal', 'In_relationship','Education_status']
data_clustered = data.copy()

model = KMeans(n_clusters=4, random_state = 7)
model.fit(data_clustered[cols_for_clustering])
data_clustered['Cluster'] = model.predict(data_clustered[cols_for_clustering])
data_clustered.head()
```

AVERAGE INCOME OF EACH CLUSTER AND BAR PLOT VISUALIZATION

```python
data_clustered['Avg_income'] = data_clustered.groupby('Cluster')['Income'].mean()

plt.figure(figsize=(8, 6))
sns.barplot(x='Cluster', y='Avg_income', data=data_clustered, palette='viridis')
plt.title('Cluster Avg_income')
plt.xlabel('Clusters')
plt.ylabel('Avg_income')
plt.show()
```

AVERAGE PRODUCT CONSUMPTION PER CLUSTER

```python
Avg_Consumption = data_clustered.groupby('Cluster')[['MntWines', 'MntFruits', 'MntMeatProducts','MntFishProducts','MntSweetProducts','MntGoldProds','MntTotal']].mean()

print(Avg_Consumption)
```

RESTRUCTING THE DATA USING MELT FUNCTION AND VISUALIZATION OF AVGRAGE CONSUMPTION

```python
Avg_Consumption["Cluster"] = [0, 1, 2, 3]  

melted_data = pd.melt(
    Avg_Consumption, 
    id_vars="Cluster", 
    var_name="Product", 
    value_name="Consumption"
)

plt.figure(figsize=(12, 10))
sns.barplot(x="Cluster", y="Consumption", hue="Product", data=melted_data, palette="viridis", errorbar=None)
plt.title("Product Consumption by Cluster")
plt.xlabel("Cluster")
plt.ylabel("Consumption")
plt.xticks(rotation=0)
plt.legend(title="Product", loc="upper right")
plt.show()
```
